Ok, now I'd like to do some test redistricting with my sample fairfax counties.

```r
library(sf)
library(ggplot2)
library(tidyverse)
library(tibble)
library(magrittr)

df <- st_read("C:/Users/madie/OneDrive/data/pre-redist/VA_precinct_2015_fairfax/VA_precinct_2015_fairfax.shp")

ggplot(data = df) +
    geom_sf() #+
```

```r
library(redist)
library(sp)
library(spData)
library(spdep)
library(igraph)

adjlist <- redist.adjacency(df)

alg_mcmc <- redist.mcmc(adjobj = adjlist,
                        popvec = df$pop,
                        ndists = 10,
                        nsims = 10000)
```

```r
# Let's plot this!
library(randomcoloR)
colors <- distinctColorPalette(k=10)

cds <- alg_mcmc$partitions[,3]
plot <- redist.map(shp=df, 
        district_membership=cds, 
        centroids=FALSE, 
        edges=FALSE,
        title="10 Districts in Fairfax County: 2015") +
        theme(legend.position="right") +
        scale_fill_manual(values = colors)
        # geom_sf_label(aes(label = cds))
#ggsave("fairfax_10.png")
plot
```

```r
ptm <- proc.time()
mcmc.out.st.multiple <- redist.mcmc(adjobj = adjlist,
                                    popvec = df$pop,
                                    ndists = 10,
                                    nsims = 100000,
                                    constraint = c("population", "compact"),
                                    constraintweights = c(5.4, 3),
                                    ssdmat = dist,
                                    temper = TRUE,
                                    maxiterrsg = 10000)
proc.time() - ptm
```

```r
colors <- distinctColorPalette(k=10)

cds <- mcmc.out.st.multiple$partitions[,1]
plot <- redist.map(shp=df, 
        district_membership=cds, 
        centroids=FALSE, 
        edges=FALSE,
        title="10 Districts in Fairfax County: Compactness Constraints: 2015") +
        theme(legend.position="right") +
        scale_fill_manual(values = colors)
        # geom_sf_label(aes(label = cds))
#ggsave("fairfax_10.png")
plot
```
