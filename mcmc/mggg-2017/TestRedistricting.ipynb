```r
## Load data
library(redist)
library(sf)
library(sp)
library(spdep)
library(igraph)
library(ggplot2)
library(dplyr)
library(magrittr)
data("algdat.pfull")
data("fl25")

## Run the simulations
mcmc.out <- redist.mcmc(adjobj = algdat.pfull$adjlist,
                        popvec = algdat.pfull$precinct.data$pop,
                        nsims = 10000,
                        ndists = 3)
```

```r
fl25 %>% ggplot(aes(fill = pop)) + 
  geom_sf() +
  ggtitle("Population")

fl25 %>% ggplot(aes(fill = vap)) + 
  geom_sf() +
  ggtitle("Voting-Age Population")

fl25 %>% ggplot(aes(fill = obama)) + 
  geom_sf() +
  ggtitle("Obama Vote")
```

```r


fl25 %>% ggplot(aes(fill = obama)) + 
    geom_sf() +
    scale_colour_gradient2(
        low = "red",
        mid = "white",
        high = "blue",
        midpoint = 0,
        space = "Lab",
        na.value = "grey50",
        guide = "colourbar",
        aesthetics = "colour")
```
