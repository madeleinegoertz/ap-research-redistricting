```r
library(randomcoloR)
colors <- distinctColorPalette(k=11)
colors
```

```r
mcmc.out.st$partitions %>%
  as_tibble() %>%
  mutate(row_n = row_number()) %>%
  rowwise() %>%
  mutate(p_cd = mean(c_across(V1:V10000))) %>%
  nest(cols = starts_with("V")) %>%
  inner_join(df, by = "row_n") %>%
  st_as_sf() %>%
  ggplot(aes(fill = p_cd)) +
  geom_sf(size = 0) +
  scale_fill_gradientn(colors = colors) +
  coord_sf(datum = NA) +
  theme_void() + 
  labs(
    title = "Hypothetical districts using a North-South dividing line",
    fill = "Assignment\nprobability"
  )
```
