Now, using `maup`, I need to disaggregate the block level population data and then reaggregate to the block level. aggregate the block group population data into the precinct file.

```python
blocks = geopandas.read_file("zip://C:/Users/madie/OneDrive/data/ipums/VA_block_2010_pop.zip")
```

```python
import maup
import warnings

# turn off annoying geoseries isna warnings
warnings.filterwarnings('ignore', 'GeoSeries.isna', UserWarning)
# enable progress bar
maup.progress.enabled = True

# change crs of shp files to be projected (using VA north)
blocks = blocks.to_crs(epsg=2283)
bg = bg.to_crs(epsg=2283)

# remove any bowties (little imperfections in the polygons)
blocks.geometry = blocks.buffer(0)
bg.geometry = bg.buffer(0)

blocks = blocks.reset_index(drop = True)
bg = bg.reset_index(drop = True)
```

```python
# assign blocks to block groups
pop_cols = ["pop"]
assignment = maup.assign(blocks, bg)
# We prorate the population totals according to each block's share of the overall
# block group population:
denom = assignment.map(bg["pop"])
weights = blocks["pop"] / denom
prorated = maup.prorate(assignment, bg[pop_cols], weights)

# Add the prorated vote totals as columns on the `blocks` GeoDataFrame:
blocks[pop_cols] = prorated
blocks.to_file("C:/Users/madie/OneDrive/data/ipums/VA_block_2018_data/VA_block_2018_data.shp")
```

```python
# cols to aggregate from blocks to precincts
variables = ['pop']
with maup.progress():
    assignment = maup.assign(bg, precincts)
precincts[variables] = bg[variables].groupby(assignment).sum()
```

Now that I've disaggregated by prorating the population data from the block groups down to blocks, I should be able to aggregate the population data from the blocks to the precincts.
