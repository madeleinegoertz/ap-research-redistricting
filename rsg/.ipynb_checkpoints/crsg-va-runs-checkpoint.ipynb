```r
library(sf)
library(ggplot2)
library(tidyverse)
library(tibble)
library(magrittr)
library(redist)
library(sp)
library(spData)
library(spdep)
library(igraph)
```

```r
run_crsg <- function(fname, year, x) {
    df <- st_read(fname)
    adj <- redist.adjacency(df)
    area <- sf::st_area(df)
    centers <- sf::st_coordinates(sf::st_centroid(df))
    
    for (i in 1:x) {
        start.time <- Sys.time()
        crsg_out <- redist.crsg(adj.list = adj, 
                                area = area, 
                                population = df$pop, 
                                x_center = centers[,1], 
                                y_center = centers[,2], 
                                ndists = 11, 
                                thresh = .1)
        end.time <- Sys.time()
        print(end.time - start.time)

        save(crsg_out, file = paste0("crsg_vac_", toString(year), "_0.1_", toString(i), ".RData"))
    }
}
```
